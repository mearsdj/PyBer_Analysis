```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
#creating a plot with ax.plot
fix, ax = plt.subplots()
ax.plot(x_axis,y_axis)
```

```python
#creating a plot using ax.plt()
fig =  plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
#create the plot using ax.plt()
ax = plt.axes()
ax.plot(x_axis,y_axis)
```

```python
#plt.show() should only be used once per session because
#it can cause the graph to display improperly.
# If you have problems with data not being displayed properly,
# you may have to restart your kernel and clear your output.

plt.plot(x_axis,y_axis)
plt.show()
```

```python
# # Create the plot and add a label for the legend.
# plt.plot(x_axis, y_axis, label='Boston')
# # Create labels for the x and y axes.
# plt.xlabel("Date")
# plt.ylabel("Fare($)")
# # Set the y limit between 0 and 45.
# plt.ylim(0, 45)
# # Create a title.
# plt.title("PyBer Fare by Month")
# # Add the legend.
# plt.legend()

plt.plot(x_axis,y_axis,marker='D',color='purple',linewidth=1,label='Boston')
plt.xlabel('Date')
plt.ylabel('Fare($)')
plt.ylim(0,45)
plt.title('PyBer Fare by Month')
plt.grid()
plt.legend()
plt.xlim(1,6)
```

```python
#redoing the plot using obj oriented method
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(x_axis,y_axis, label='Boston', marker='D',color='purple',linewidth=1)
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
ax.set_title('PyBer Fare by Month')
ax.legend()
ax.grid()
#ax.set_xlim(0,5)
    
```

```python
#pie charts!
plt.pie(y_axis,labels=x_axis)  
```

```python
#adjust plot size so percentages fit better
plt.subplots(figsize=(8,8))
explode_values = tuple(0.2 if x==max(y_axis) else 0 for x in y_axis)
plt.pie(y_axis, labels=x_axis, explode=explode_values,autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = tuple(0.2 if x==max(y_axis) else 0 for x in y_axis)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()  
```

```python
#creating same pie using obj oriented approach
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis,labels=x_axis)
plt.show()
```

```python
#creating fully formatted pie using obj oriented approach
fig, ax = plt.subplots(figsize=(8,8))
second_largest=list(set(y_axis))
second_largest.sort()
second_largest=second_largest[-2]

explode_values = tuple(0.2 if (x==max(y_axis) or x == second_largest) else 0 for x in y_axis)

ax.pie(y_axis,labels=x_axis, colors=colors,explode=explode_values, autopct='%.1f%%', shadow=True,startangle=270)
ax.invert_yaxis()
plt.show()
```
